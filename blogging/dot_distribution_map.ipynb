```python
import folium
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv')
data = df.loc[(df['yr_built'] >= 2000) & (df['yr_built'] < 2010)]
data.head()
```

```python
m = folium.Map(location=[47.5480,-121.9836], tiles="Mapbox Bright", zoom_start=10)
```

```python
# I can add marker one by one on the map
for i in range(0,len(data)):
   folium.Circle(
      location=[data.iloc[i]['lat'], data.iloc[i]['long']],
      radius=1, #data.iloc[i]['price']*2,
      color='crimson',
      fill=True,
      fill_color='crimson',).add_to(m)
# save it as html
m.save('2000s_map.html')
```

```python
data = df.loc[(df['yr_built'] >= 1990) & (df['yr_built'] < 2000)]
data.head()

for i in range(0,len(data)):
   folium.Circle(
      location=[data.iloc[i]['lat'], data.iloc[i]['long']],
      radius=1, #data.iloc[i]['price']*2,
      color='blue',
      fill=True,
      fill_color='crimson',).add_to(m)
```

```python
# Save it as html
m.save('1990s_vs_2000s_map.html')
```
